# Terminologies:
Consider two functions  f(n)  and  g(n)  that are defined for all positive integers and take on non-negative real values. We say that f grows slower than g and write f < g as n grows.


Now this can be classified into two groups:

    (i) if f < g then   f=o(g)  (small-o).
    (ii) if f <=g then f=O(g)  (Big - O).

In this notebook let us keep our interest on *Big- O* notation only. Now a common misconception arises that $5n^2=O(n^3)$ is not true. But it is absolutely valid. Many say that $5n^2 = O(n^2)$ which is correct. But both of them are 100% correct. Like $2<3$ doesn't mean that 2 is not less than 4. If $2<3$ then $2<4$ and like that $5n^2<=n^2$ also means $5n^2<n^3$. Thus both of them are correct.

# Some Rules

    (i) Multiplicative constants can be omitted:  c.f <= f. Like 5n^2<=n^2
    (ii) Out of two functions the one with higher degree grows faster.
    (iii) Any polynomial grows slower than any exponential. n^a < b^n (a>=0, b>1)
    (iv) Smaller terms can be ignored. For eg., n^2 + n <= n^2

Enough of this theoretical concepts. Let us visualize some of them...

#### A naive sequential search|| $O(n^2)$


```python
# instantiate arrays containing both negative and positive values
true_array = np.sort([np.random.randint(-20, 20) for i in range(20)]) # create a random sorted array containing both negative and positive values
false_array = np.sort([-4, -2, 1, 3, 5, 7, 9])

print("Non negative array: ", "\n", true_array, "\n", "Array with negative values: ", "\n", false_array)
```

```python
def naive_search(arr):
  """
  Naive sequential search of a given array
  """
  b_arr = []

  for a in arr:
    if a < 0:
      for b in arr:
        if b == abs(a):
          b_arr.append(a)
          
  if b_arr:
    return f"{True}, the array is:  {b_arr}"
  else:
    return False
```

```python
#test naive sequential search
print(naive_search(true_array) )
print("\n")
print(naive_search(false_array) )
```

```python
def seq_search(arr):
  """
  A linear sequential search for the negative of any elements of the given array.
  arr[numpy array OR python list]: an array containing lists of random negative and non-negative integers.
  """
  arr = np.sort(arr) # sort array

  array = []
  for a in arr:
    if a*(-1) in arr:
      array.append(a)
  
  if len(array) !=0:
    return f"{True}, \nThe array is:  {array}"
  else:
    return False    
  
```

```python
%%file functionprofiles.py
from memory_profiler import profile
import numpy as np
import typer 
app = typer.Typer()

@app.command()
def create_sparse_matrix(dim, num):
  """
  Function creates a sparse matrix
  """
  ## initialize a zero matrix
  sparse_initial = np.zeros((dim, dim))
  np.random.seed(10)
  val_update =  np.array([np.random.randint(10) for i in range(num)])
  for idx, val in enumerate(val_update):
    i = np.random.randint(0, dim)
    j = np.random.randint(0, dim)
    sparse_initial[i][j] = val

  return sparse_initial

def sparse_representation(mat):
  import numpy
  """
  computes the index-based representation of a sparse matrix
  Reference: https://www.geeksforgeeks.org/sparse-matrix-representation/
  """
  # initialize size
  size = 0  
  mat = numpy.array(mat)
  row, column = mat.shape
  for i in range(row):
    for j in range(column):
      if (mat[i][j] != 0):
        size+=1

  # the index-based representation of the matrix should be equal to the number 
  # of non-zero elements in the input matrix
  rows, cols = (3, size)
  index_mat = [[0 for i in range(cols)] for j in range(rows)]
  k = 0

  for i in range(row):
    for j in range(column):
      if (mat[i][j] != 0):
        index_mat[0][k] = i
        index_mat[1][k] = j
        index_mat[2][k] = mat[i][j]
        k+=1      
  return index_mat


# SUMMATION FUNCTIONS
@app.command()
@profile(precision=4)
def matrix_sum(dim=10, num=4):
  """
  Matrix summation function
  """
  mat = create_sparse_matrix(dim, num)
  row = 0
  for column in mat:
    for rows in column:
      row+=rows
  
  return row


@app.command()
@profile(precision=4)
def sparse_mat_sum(dim=10, num=1):
  """
  Computes the sum of the Row, Column, Value representation of a sparse matrix
  """
  mat = create_sparse_matrix(dim, num)
  sparse_mat = sparse_representation(mat)
  for idx, val in enumerate(sparse_mat):
    if idx > 1:
      return (np.sum(val))
```

```python
# PAY ATTENTION TO THE PERCENTAGE OF NULL VALUES IN THE SPARSE MATRICES
# different dimensions of matrices
# simple direct sparse matrix
def create_sparse_matrix(dim, num):
  """
  Function creates a sparse matrix
  """
  ## initialize a zero matrix
  sparse_initial = np.zeros((dim, dim))
  np.random.seed(10)
  val_update =  np.array([np.random.randint(10) for i in range(num)])
  for idx, val in enumerate(val_update):
    i = np.random.randint(0, dim)
    j = np.random.randint(0, dim)
    sparse_initial[i][j] = val
  return sparse_initial


def product_mat(A, B):
  product = np.zeros((A.shape[0], B.shape[1]))
  for a in range(A.shape[0]):
    for b in range(B.shape[1]):
      for c in range(A.shape[1]):
        product[a][b] += A[a][b] * B[a][b]
  return product

def matrix_multiplication(A, B):
    return np.dot(A, B)
```

```python
val_range = np.arange(10, 1100, 100)
first_matrices = [create_sparse_matrix(d, int(0.2*(d*d)) ) for d in val_range]
second_matrices = [create_sparse_matrix(d, int(0.2*(d*d)) ) for d in val_range]
```

```python
time_elapse_loop = []

for A, B in zip(first_matrices, second_matrices):
  start = time.time()
  product_mat(A, B)
  stop = time.time()
  time_elapse_loop.append((stop - start))
```

```python
time_elapse_matrix = []

for A, B in zip(first_matrices, second_matrices):
  start = time.time()
  matrix_multiplication(A, B)
  stop = time.time()
  time_elapse_matrix.append((stop - start))
```
